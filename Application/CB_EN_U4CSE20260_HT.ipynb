**H0:mu=59.8725**

**H1:mu!=59.8725**




alpha=0.05

n=50

```python
import numpy as np 
import scipy.stats as st
import pandas as pd
```

```python
df=pd.read_csv("/content/songs_normalize.csv")
df
```

```python
df.describe()
```

alpha =alpha/2 two tailed test

```python
mu=np.mean(df['popularity'])
mu
```

```python
sigma=np.std(df['popularity'])
sigma
```

```python
n=50
alpha=0.025
```

```python
samp=df.sample(50)
samp
```

```python
xbar=np.mean(samp['popularity'])
xbar
```

**Z Test**

```python
z_critical = abs(st.norm.ppf(alpha))
z_critical
```

The critical values are -1.96 and 1.96 since this is a two tailed test

```python
z=(xbar-mu)/(sigma/np.sqrt(n))
z
```

```python
if z>=z_critical or z<=-z_critical:
  print("reject null hypothesis")
else:
  print("Null hypothesis cannot be rejected")
```

Therefore the we cannot reject null hypothesis and the average popularity in the spotify dataset is 59.8725

**T test**

Now consider a sample size of 10

```python
n = 10
dof = n-1
```

```python
samp1=df.sample(10)
samp1
```

```python
xbar=np.mean(samp1['popularity'])
xbar
```

```python
t = (xbar - mu)/(sigma/ np.sqrt(n))
t
```

```python
t_critical = st.t.ppf(alpha, dof)
t_critical
```

```python
if (abs(t) > abs(t_critical)): #Absolute value taken as the it's a two-tailed test and the original t_critical value might be negative
    print("Reject null hypothesis")
else:
    
    print("Null hypothesis cannot be rejected")
```

Hence the null hypothesis cannot be rejected. Thus from both the test we can conclude mu=59.8725 and we fail to reject the null hypothesis.
